```python
import pytesseract
import os
from PIL import Image
from pdf2image import convert_from_path
import pandas as pd
import re
import numpy as np
from IPython.display import display
import cv2
from itertools import zip_longest
```

```python
pytesseract.pytesseract.tesseract_cmd = r'D:\tesseract\tesseract.exe'
poppler_path = r'D:\poppler-24.08.0\Library\bin'
config = r'--psm 6 -l eng --oem 3 tessedit_char_whitelist=ABCDEFG0123456789'
```

```python
df = pd.DataFrame(columns = ['Post Date', 'Description', 'Debits', 'Credits', 'Balance'])
```

```python
filePath = '_.pdf'

doc = convert_from_path(filePath, poppler_path = poppler_path)
path, fileName = os.path.split(filePath)
fileBaseName, fileExtension = os.path.splitext(fileName)

for page_number, page_data in enumerate(doc):
    page_data = np.array(page_data)
    gray = cv2.cvtColor(page_data, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    thresh = cv2.GaussianBlur(thresh, (3,3), 0)

    d = pytesseract.image_to_data(thresh, output_type= 'dict')
    n_boxes = len(d['level'])
    prev_balance = 0
    for i in range(n_boxes):
        (x, y, w, h, text) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i], d['text'][i])
        # Parsing table
        if text in ['Post']:
            im = Image.fromarray(page_data)
            date_crop = im.crop((x, y + 35, x + w + 100, y + h + 2000))
            desc_crop = im.crop((x + 150, y + 35, x + w + 800, y + h + 2000))
            credit_or_debit_crop = im.crop((x + 850, y + 35, x + w + 1300, y + h + 2000))
            balance_crop = im.crop((x + 1300, y + 35, x + w + 1700, y + h + 2000))
            
            date_text = pytesseract.image_to_string(date_crop, config = config)
            description_text = pytesseract.image_to_string(desc_crop, config = config)
            credit_or_debit_text = pytesseract.image_to_string(credit_or_debit_crop, config = config)
            balance_text = pytesseract.image_to_string(balance_crop, config = config)

            date_text_list = list(filter(lambda x: re.match(r"(\d{2})/(\d{2})/(\d{4})", x), date_text.split('\n')))
            credit_or_debit_text_list = list(filter(lambda x: x != '', credit_or_debit_text.split('\n')))
            description_text_list = list(filter(lambda x: x != '' and '>' not in x, description_text.split('\n')))
            balance_text_list = list(filter(lambda x: x != '', balance_text.split('\n')))
            if description_text_list[0].startswith('Beginning'):
                credit_or_debit_text_list.insert(0, None)
            
            completed_description = []
            for idx in range(len(description_text_list)):
                current_element = description_text_list[idx]
                if (description_text_list[idx - 1].endswith('*') or re.search(r'[:/]', description_text_list[idx - 1]))\
                    and len(completed_description) != 0:
                    last_element = completed_description.pop()
                    element = last_element + ' ' + current_element
                    completed_description.append(element)
                else:
                    completed_description.append(current_element)

            # print(len(date_text_list), len(description_text_list), len(credit_or_debit_text_list), len(balance_text_list))
            zipped_data = list(zip_longest(date_text_list, completed_description, credit_or_debit_text_list, balance_text_list, fillvalue = None))
            for row in zipped_data:
                date = row[0]
                desc = row[1]
                credit_or_debit = float(re.sub('[$,]', '', row[2])) if row[2] is not None else None
                balance = float(re.sub('[$,]', '', row[3]))
            
                # assumption that first row is debit (if prev_balance = +inf -> debit)
                if desc.startswith('Beginning'):
                    prev_balance = balance
                    
                if balance < prev_balance:
                    # [Post Date, Description, Debit, Credit, Balance]
                    df.loc[len(df)] = [date, desc, credit_or_debit, None, balance]
                else:
                    df.loc[len(df)] = [date, desc, None, credit_or_debit, balance]
                prev_balance = balance

        # Parsing account info
        if text in ['Type']:
            im = Image.fromarray(page_data)
            account_type_crop = im.crop((x - 120, y + 30, x + w + 400, y + h + 50))
            account_number_crop = im.crop((x + 500, y + 30, x + w + 1000, y + h + 50))

            account_type_text = pytesseract.image_to_string(account_type_crop, config = config)
            account_number_text = pytesseract.image_to_string(account_number_crop, config = config)

            if account_type_text.endswith('\n'):
                account_type_text = account_type_text[:-2]
            if account_number_text.endswith('\n'):
                account_number_text = account_number_text[:-2]

            df.attrs['account_type'] = account_type_text
            df.attrs['account_number'] = account_number_text
```
